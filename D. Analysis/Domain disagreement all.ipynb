```python
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define data manually as a dictionary
data = {
    'Domain_EN': [
        'Law & Justice', 'Politics & Leadership', 'Religion & Spirituality', 'Science & Technology',
        'Journalism & Media', 'Education', 'Hospitality & Service', 'Activism & Social Change',
        'Caregiving & Parenting', 'Healthcare', 'Arts & Culture', 'Business & Economics'
    ],
    'GPT-4':                [5, 4, 4, 3, 3, 2, 2, 2, 2, 1, 1, 1],
    'LLaMA3.1':             [4, 2, 2, 2, 4, 2, 3, 2, 2, 1, 1, 3],
    'LLaMA3.2':             [5, 5, 4, 3, 4, 3, 4, 5, 3, 4, 4, 2],
    'Qwen2.5':              [4, 3, 3, 2, 3, 2, 4, 2, 2, 3, 4, 3],
    'mBERT':               [5, 5, 4, 5, 5, 5, 5, 5, 5, 5, 4, 5],
    'XLM-R':               [5, 5, 5, 4, 5, 4, 4, 5, 5, 5, 5, 3],
}

# Create dataframe
df = pd.DataFrame(data)

# Melt for seaborn
df_melt = df.melt(id_vars='Domain_EN', var_name='Model', value_name='Disagreements')

# Sort domains for consistency
df_melt['Domain_EN'] = pd.Categorical(df_melt['Domain_EN'], categories=data['Domain_EN'], ordered=True)

# Define color palette (logo colors or similar)
model_colors = {
    'GPT-4': '#9b59b6',
    'LLaMA3.1': '#3498db',
    'LLaMA3.2': '#2980b9',
    'Qwen2.5': '#f39c12',
    'mBERT': '#2ecc71',
    'XLM-R': '#e74c3c',
}

# Plot
plt.figure(figsize=(14, 6))
sns.barplot(data=df_melt, x='Domain_EN', y='Disagreements', hue='Model', palette=model_colors)

plt.xticks(rotation=45, ha='right')
plt.title('🌍 Domain-wise Disagreements Across LLMs')
plt.xlabel('Domain')
plt.ylabel('Disagreement Count')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Save
# Save to PDF
plt.savefig('domain_disagreements.pdf', format='pdf')

# Show plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Model disagreement data
data = {
    'Domain_EN': [
        'Law & Justice', 'Politics & Leadership', 'Religion & Spirituality', 'Science & Technology',
        'Journalism & Media', 'Education', 'Hospitality & Service', 'Activism & Social Change',
        'Caregiving & Parenting', 'Healthcare', 'Arts & Culture', 'Business & Economics'
    ],
    'GPT-4':             [5, 4, 4, 3, 3, 2, 2, 2, 2, 1, 1, 1],
    'LLaMA3.1':          [4, 2, 2, 2, 4, 2, 3, 2, 2, 1, 1, 3],
    'LLaMA3.2':          [5, 5, 4, 3, 4, 3, 4, 5, 3, 4, 4, 2],
    'Qwen2.5':           [4, 3, 3, 2, 3, 2, 4, 2, 2, 3, 4, 3],
    'mBERT':             [5, 5, 4, 5, 5, 5, 5, 5, 5, 5, 4, 5],
    'XLM-Roberta':       [5, 5, 5, 4, 5, 4, 4, 5, 5, 5, 5, 3]
}

# Create DataFrame
df = pd.DataFrame(data)
df = df.set_index('Domain_EN')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.T, annot=True, cmap="Greens", cbar=True, linewidths=0.5, linecolor='gray')

# Styling
plt.title('Heatmap of Domain-Level Disagreements Across Models', fontsize=14)
plt.xlabel('Domain')
plt.ylabel('Model')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Save as PDF
plt.savefig('heatmap_domain_disagreements.pdf', format='pdf')

# Show plot
plt.show()
```
